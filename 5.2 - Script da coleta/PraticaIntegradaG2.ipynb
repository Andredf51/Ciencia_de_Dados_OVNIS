1° Base de dados

```
#Base de dados: http://www.nuforc.org/
#Extração de dados de forma tabular de vinte anos (Set/1997 - Agosto/2017 = 240 pág)
 
import pandas as pd
from bs4 import BeautifulSoup
import requests
import time
 
#Teste de requisição HTTP
requisicao = requests.get('http://www.nuforc.org/webreports/ndxevent.html')
requisicao
```

```
#Teste de importação de dados com 1 página (01/2021)
dfs = pd.read_html('http://www.nuforc.org/webreports/ndxe202101.html')
df = dfs[0]
#mostrar os 5 primeiros
df.head()
#Contar total de relatos
#Contar as linhas do dataframe df
total_rows = df.count
print(total_rows)
 
#Obs: URL do site - http://www.nuforc.org/webreports/ndxe +
# data - ano mês +
#html
```

```
#lista com os dados
lista10 = []

#for para percorrer as páginas
#range (início, fim, intervalo)
mes=0
 #Ano 1997
for mes in range (9,13,1):
        if mes<10:
          
          data=str(1997)+str(0)+str(mes)            
          if requisicao.status_code == 200:              
              lista10.append(f'http://www.nuforc.org/webreports/ndxe{data}.html')           
          
          
          
        else:
          data=str(1997)+str(mes)          
          if requisicao.status_code == 200:
              
              lista10.append(f'http://www.nuforc.org/webreports/ndxe{data}.html')
time.sleep(1)


#Quantidade de páginas na lista
len(lista10)
```

```
#Anos 1998 até 2016
lista11 = []
for ano in range (1998,2017,1):
  #print(f'Página: {ano}')
  mes=0
  data=0
  for mes in range (1,13,1):
      if mes<10:
          
        data=str(ano)+str(0)+str(mes)            
        if requisicao.status_code == 200:              
            lista11.append(f'http://www.nuforc.org/webreports/ndxe{data}.html')           
          
          
          
      else:
        data=str(ano)+str(mes)          
        if requisicao.status_code == 200:
              
            lista11.append(f'http://www.nuforc.org/webreports/ndxe{data}.html')

  time.sleep(1)

#Quantidade de páginas na lista
len(lista11)
```

```
#Ano 2017
lista12 = []

#for para percorrer as páginas
#range (início, fim, intervalo)
mes=0
 #Ano 1997
for mes in range (1,9,1):
        if mes<10:
          
          data=str(2017)+str(0)+str(mes)            
          if requisicao.status_code == 200:              
              lista12.append(f'http://www.nuforc.org/webreports/ndxe{data}.html')           
          
          
          
        else:
          data=str(ano)+str(mes)          
          if requisicao.status_code == 200:
              
              lista12.append(f'http://www.nuforc.org/webreports/ndxe{data}.html')
time.sleep(1)

#Quantidade de páginas na lista
len(lista12)
```

```
#Criar uma lista das listas anteriores
#lista nova para criaçâo de tabelas - lista2
#Variável conte - controle de listas
# reunir todas as lista em uma (lista2)
# lista 2 possui todos os relados por mês

lista2 = []
for conte in lista10:
  pag = pd.read_html(conte, header = 0)
  lista2.append(pag[0])

lista3 = []
for conte in lista11:
  pag = pd.read_html(conte, header = 0)
  lista3.append(pag[0])

lista4 = []
for conte in lista12:
  pag = pd.read_html(conte, header = 0)
  lista4.append(pag[0])

lista2.append(lista4)
lista2.append(lista3)
lista2
```

```
#Criação do dataFrame

df_relatos = pd.DataFrame()
for i in lista2:
  df_relatos = df_relatos.append(i).reset_index(drop = True)


df_relatos
```

3° Salvar em um arquivo .csv (OVNIS.csv)

```
#Código para salvar o dataFrame 
df_relatos.to_csv('OVNIS.csv')
```
